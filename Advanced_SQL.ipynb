Q1. What is a Common Table Expression (CTE), and how does it improve SQL query readability?



A CTE is a temporary result set created using the WITH keyword.
It helps make complex queries easier to read and understand by breaking them into smaller parts.

It improves readability because the query becomes more organized.

Q2. Why are some views updatable while others are read-only? Explain with an example.



**->**    A view is updatable when it is created from a single table without using joins or aggregate functions.

A view becomes read-only when it contains:

* GROUP BY
* DISTINCT
* Aggregate functions
* Joins

**Updatable View Example**:

CREATE VIEW ProductView AS

SELECT ProductID, ProductName, Price

FROM Products;

**Read-only View Example**:

CREATE VIEW AvgPriceView AS

SELECT Category, AVG(Price)

FROM Products

GROUP BY Category;


Q3. What advantages do stored procedures offer compared to writing raw SQL queries repeatedly?



**->**   Advantages of stored procedures:

* **Code Reusability**

Write once and execute many times.

* **Better Performance**

Procedures are precompiled by MySQL.

* **Improved Security**

Users can execute procedures without direct table access.

 * **Reduced Network Traffic**

Only procedure call is sent instead of long SQL queries.

   * **Easy Maintenance**

Business logic is centralized.


Q4. What is the purpose of triggers in a database? Mention one use case where a trigger is essential.



**->** A trigger is a database object that runs automatically when an INSERT, UPDATE, or DELETE operation happens.

**Purpose:**
* Maintain logs
* Apply business rules
* Automatically update data

**Example Use Case:**

When a product is deleted, its details can be stored in another table for backup.

Q5. Explain the need for data modelling and normalization when designing a database.

**Data modeling** helps organize data properly in tables and relationships.

**Normalization**  helps remove duplicate data and improves data consistency.

Benefits:
* Reduces redundancy

* Saves storage space

* Improves data integrity

* Makes database easier to manage

**Example:**

Instead of storing customer details repeatedly in every order, we store them in a separate customer table.
